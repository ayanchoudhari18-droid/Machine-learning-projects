```python
def fibonacci_series(n_terms):
    a, b = 0, 1
    count = 0

    if n_terms <= 0:
        print("Please enter a positive integer")
    elif n_terms == 1:
        print("Fibonacci series up to", n_terms, ":")
        print(a)
    else:
        print("Fibonacci series up to", n_terms, ":")
        while count < n_terms:
            print(a)
            nth = a + b
            a = b
            b = nth
            count += 1

# Example usage: print the first 10 terms of the Fibonacci series
fibonacci_series(10)
```
