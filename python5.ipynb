```python
day=int(input("Enter a day:"))
match day:
    case 1:
        print("SUNDAY")
    case 2:
        print("MONDAY")
    case 3:
        print("TUSESDAY")
    case 4:
        print("WEDNESDAY")
    case 5:
        print("THURSDAY")
    case 6:
       print("FRIDAY")
    case 7:
        print("SATURDAY")
    case _:
        print("INVALALID DAY")
```
