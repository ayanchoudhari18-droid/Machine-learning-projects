```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

#STUDY HOURS(input)
X=np.array([1,2,3,4,5]).reshape(-1,1)

#result (0=fail,1=pass)
y=np.array([0,0,1,1,1])

#create model
model=LogisticRegression()
model.fit(X,y)

#create smooth curve
X_test=np.linspace(0,6,100).reshape(-1,1)
y_prob=model.predict_proba(X_test)[:, 1]

#plot data points
plt.scatter(X,y)

#plot smooth curve
plt.plot(X_test,y_prob)

#label
plt.xlabel("study hours")
plt.ylabel("probability of pass")
plt.title("logistic regression:pass/fail prediction")
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

x=np.array([[1],[2],[3],[4],[5],[6],[7],[8]])
y=np.array([0,0,0,1,1,1,1,1])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)

model=LogisticRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)

print("Actual:",y_test)
print("Predicted:",y_pred)
print("accuracy:",accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

x=np.array([
    [4,0], #Basic
    [5,1],  #Basic
    [6,1],  #Basic
    [10,3], #Primium
    [12,4], #Primium
    [14,5], #Primium
    [9,2],#Primium
    [5,0] #Basic
])

y=np.array([0,0,0,1,1,1,1,0])

model=KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)
new_customer = np.array([[11,3]])
prediction = model.predict(new_customer)

print("prediction membership :", "Primium" if prediction[0] == 0 else "Basic")

for i in range(len(x)):
  if y[i]==0:
    plt.scatter(x[i][0],x[i][1],color='blue',label="Basic" if i==0 else "")
  else:
    plt.scatter(x[i][0],x[i][1],color='red',label="Primium" if i==3 else "")

plt.scatter(new_customer[0][0],new_customer[0][1],color='green',s=200,label='New custmer')
plt.xlabel("Monthly visits")
plt.ylabel("training sessions")
plt.title("KNN membership classification")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x=np.array([
    [22,2000],
    [23,2200],
    [23,2100],
    [45,8000],
    [50,8500],
    [48,8200],
])

model=KMeans(n_clusters=2)
model.fit(x)
labels=model.labels_
centers=model.cluster_centers_

for i in range(len(x)):
  if labels[i]==0:
    plt.scatter(x[i][0],x[i][1],color='blue')
  else:
    plt.scatter(x[i][0],x[i][1],color='red')

plt.scatter(centers[:,0],centers[:,1],color='green',s=200,label='center')
plt.xlabel("Age")
plt.ylabel("Monthly spending")
plt.title("customer clustring using k-means")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
x=np.array([
    [42,50],
    [38,43],
    [85,88],
    [90,92],
    [87,85],
    [60,65],
    [62,63]
])
inertia=[]

for k in range(1,6):
  kmeans=KMeans(n_clusters=k,random_state=42, n_init=10)
  kmeans.fit(x)
  inertia.append(kmeans.inertia_)

plt.figure()
plt.plot(range(1,6),inertia,marker='o')
plt.xlabel("number of clusters")
plt.ylabel("inertia")
plt.title("Elbow method")
plt.show()




model=KMeans(n_clusters=3,random_state=42)
model.fit(x)

labels=model.labels_
centers=model.cluster_centers_

colors=['blue','red','green']

plt.figure()
for i in range(len(x)):
    plt.scatter(x[i][0],x[i][1],color=colors[labels[i]])

plt.scatter(centers[:,0],centers[:,1],color='black',s=200)

plt.xlabel("math marks")
plt.ylabel("science marks")
plt.title("student clustring")
plt.show()




unique, counts=np.unique(labels,return_counts=True)

plt.figure()
plt.bar(unique,counts)
plt.xlabel("cluster number")
plt.ylabel("number of student")
plt.title("cluster size comparision")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
x=np.array([
    [40,45,42],
    [42,50,48],
    [38,43,40],
    [85,88,90],
    [90,92,95],
    [87,85,89],
    [60,65,63],
    [62,63,60],
])
model=KMeans(n_clusters=3,random_state=42)
model.fit(x)
labels=model.labels_
centers=model.cluster_centers_

plt.figure()
sns.heatmap(x,annot=True,cmap='coolwarm',xticklabels=["Math","Science","english"])
plt.title("student marks heatmap")
plt.show()

#3d clustering
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
for i in range(len(x)):
  ax.scatter(x[i][0],x[i][1],x[i][2],color=colors[labels[i]])

ax.scatter(centers[:,0],centers[:,1],centers[:,2],color='black',marker='x',s=200)
ax.set_xlabel("Math")
ax.set_ylabel("Science")
ax.set_zlabel("English")
ax.set_title("Student 3d clustering")
plt.show()
```
